```python
data['timestamp'] = timeStamps
```

```python
from datetime import timezone# data['timestamp'][0]
timeStampsUnix = []
for eachTime in data['timestamp']:
    timestamp = eachTime.replace(tzinfo=timezone.utc).timestamp()
    timeStampsUnix.append(int(timestamp))
    
data['timeStampUnix'] = timeStampsUnix
data.head()
```

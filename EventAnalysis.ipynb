```python
import pandas as pd 
```

```python
data = pd.read_csv("MergedDataset.csv") 
#data.head()

data = data.sort_values(by=['timestamp'])
data.head()

lt = []

current = 424.5
for index,obj in data.iterrows() :
    k = abs(obj['Weighted_Price']-current)
    if(k > 200):
        lt.append(obj['timestamp'])
       # print(obj['text'])
    current = obj['Weighted_Price'] 
        
print(lt)        
```

```python
import requests 
import time
from datetime import  datetime, date,timedelta

ltn = len(lt)
    
lt1 = lt[0:int(ltn/4)]
lt2 = lt[int(ltn/4):int(ltn/2)]
lt3 = lt[int(ltn/2):int(3*ltn/4)]
lt4 = lt[int(3*ltn/4):ltn]

for i in lt1:
    i = datetime.strptime(i, '%Y-%m-%d %H:%M:%S')
    print(i)
    j = i + timedelta(days=1)
    k=  i - timedelta(days=1)
    j = str(j)
    k = str(k)
    #p
    #URL = "https://newsapi.org/v2/everything?q=bitcoin&from="+i[0:10]+"&to="+i[0:10]+"&sortBy=popularity&apiKey=5caebbc474d042e9983eccacae604e9e"
    URL = "https://api.nytimes.com/svc/search/v2/articlesearch.json?fq=bitcoin&facet_field=day_of_week&facet=true&begin_date="+k[0:10]+"&end_date="+j[0:10]+"&api-key=1zzgk0qol54v3FXFuTfmUd5wTitTMWRS"
    r = requests.get(url = URL) 
    #print(type(i))
    data = r.json()
    #print(data['response']['docs'][0]['web_url'])
    #print(URL)
    k = data['response']['docs']
    for t in k :
        print(t['web_url'])
        


       
```

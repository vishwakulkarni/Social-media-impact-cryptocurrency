```python
import pandas
from textblob import TextBlob
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
```

```python
# Hyperparameters

window = 20
```

```python
dataset = pandas.read_csv("MergedDataset.csv")
dataset.columns = [x.lower() for x in dataset.columns]
dataset["tweet_sentiment"] = [TextBlob(text).sentiment.polarity for text in dataset["text"]]
dataset = dataset.filter(["tweet_sentiment", "close"])
```

```python
X, y = [], []
for i in range(len(dataset) - window - 1):
    row = list(dataset["tweet_sentiment"][i: i + window])
    X.append(row)
    y.append(float(dataset["close"][i + window]))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_test, y_train, y_test = np.array(X_train), np.array(X_test), np.array(y_train), np.array(y_test)
X_train, X_test = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1])), np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = Sequential()
model.add(LSTM(4, input_shape=(1, window)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X, y, epochs=100, batch_size=1, verbose=2)
```
